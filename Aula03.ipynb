# Histograma de imagens digitais
Distribuição de probabilidade que fornece informações para realce e análise de imagens, em gráfico.

Intensidade e distruição de cada pixel.
claro >< escuro

```python
import cv2
import numpy as np
import matplotlib as mtl
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("lenna.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
#gráfico do histograma da imagem
plt.hist(img.ravel(), 256, [0,255])
plt.show()
```

```python
img = cv2.imread("fusca.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap="gray") #cmap="Greys_r", vmin
```

```python
plt.hist(img.ravel(), 256,[0,255])
plt.show()
```

```python
im_eq = cv2.equalizeHist(img)
plt.imshow(im_eq, cmap="gray")
```

```python
plt.hist(im_eq.ravel(), 256,[0,255])
plt.show()
```

```python
imagem = cv2.imread("bola.png")
image = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(image, vmin=0, vmax=255); plt.show()
plt.hist(image.ravel(),256,[0,256]); plt.show()
```

```python
#histograma Vermelho
plt.imshow(image[:,:,0], cmap="gray"
, vmin=0, vmax=255); plt.show()
plt.hist(image[:,:,0].ravel(),256,[0,256]); plt.show()
```

```python
# Histograma Verde
plt.imshow(image[:,:,1], cmap="Greys_r"
, vmin=0, vmax=255); plt.show()
plt.hist(image[:,:,1].ravel(),256,[0,256]); plt.show()
```

```python
# Histograma Azul
plt.imshow(image[:,:,2], cmap="Greys_r"
, vmin=0, vmax=255); plt.show()
plt.hist(image[:,:,2].ravel(),256,[0,256]); plt.show()
```

```python
#extraindo a bolinha amarela
image2 = image.copy()
#camada 0- vermelho
#1 - verde
#2 - azul
gray_r = image2[:,:,0]
gray_g = image2[:,:,1]
gray_b = image2[:,:,2]
img_bola = image2.copy()

for y in range(0, image2.shape[0]):
  for x in range(0, image2.shape[1]):
    if gray_g[y][x] <= 230:
      img_bola[y][x]= 0
    if gray_b[y][x] >= 240:
      img_bola[y][x]= 0

plt.imshow(img_bola, interpolation="none")
plt.show()
```

```python
#extraindo a bolinha laranja
image2 = image.copy()
#camada 0- vermelho
#1 - verde
#2 - azul
gray_r = image2[:,:,0]
gray_g = image2[:,:,1]
gray_b = image2[:,:,2]
img_bola = image2.copy()

for y in range(0, image2.shape[0]):
  for x in range(0, image2.shape[1]):
    if gray_r[y][x] <= 240:
      img_bola[y][x]= 0
    if gray_b[y][x] >= 30:
      img_bola[y][x]= 0

plt.imshow(img_bola, interpolation="none")
plt.show()
```

```python
#extraindo a bolinha azul
image2 = image.copy()
#camada 0- vermelho
#1 - verde
#2 - azul
gray_r = image2[:,:,0]
gray_g = image2[:,:,1]
gray_b = image2[:,:,2]
img_bola = image2.copy()

for y in range(0, image2.shape[0]):
  for x in range(0, image2.shape[1]):
    if gray_r[y][x] >= 60:
      img_bola[y][x]= 0
    if gray_b[y][x] <= 170:
      img_bola[y][x]= 0

plt.imshow(img_bola, interpolation="none")
plt.show()
```
