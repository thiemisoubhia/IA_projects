```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
import math

I = cv2.imread('jogoMemoria.JPG', 0)
print("imagem")
plt.imshow(I, cmap='gray')
plt.show()

T = cv2.imread('laranja.png', 0)
print("template")
plt.imshow(T, cmap='gray')
plt.show()
```

```python
R = cv2.matchTemplate(I, T,cv2.TM_SQDIFF_NORMED)
limiar = 0.012

plt.imshow(R < limiar, cmap='gray')
plt.colorbar()
```

```python
I_rgb = cv2.cvtColor(I, cv2.COLOR_GRAY2BGR)
alturaT = np.size(T, 0)
larguraT = np.size(T, 1)

limiar = 0.012
locais = np.where(R < limiar)

for (x, y) in zip(*locais[::-1]):
    cv2.rectangle(I_rgb, (x, y), (x + larguraT, y + alturaT), (0,0,255), 2)

plt.imshow(I_rgb)
plt.show()
```
