# HSV
Matriz: pigmento de uma cor em graús 0 à 359

Saturação: 0 à 100

Brilho: 0 à 100

OpenCV vai de 0 à 255 (dividir por 2 para a cor)



H 0-359 (openCV 0-255) *sempre dividir por 2

S - cor mais pura (para o preto)

V - brilho (para o branco)


```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("bolinha.png")
#convertendo para a RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#convertendo para a HSV
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
#Definindo valores minimo e max de mascara
#o magenta tem h=300 mais ou menos ou 150 para OpenCV
image_lower_hsv = np.array([140, 50, 50])
image_upper_hsv = np.array([170, 255, 255])

mask_hsv = cv2.inRange(img_hsv, image_lower_hsv, image_upper_hsv)

plt.subplot()
plt.imshow(img_rgb)
plt.show()
plt.subplot()
plt.imshow(mask_hsv,cmap="Greys_r", vmin=0, vmax=255)
```

```python
img = cv2.imread("bolinha.png")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

image_lower_hsv = np.array([140, 50, 50])
image_upper_hsv = np.array([170, 255, 255])

mask_hsv = cv2.inRange(img_hsv, image_lower_hsv, image_upper_hsv)

plt.subplot()
plt.imshow(img_rgb)
plt.show()
plt.subplot()
plt.imshow(mask_hsv,cmap="Greys_r", vmin=0, vmax=255)
```

```python
#reduzir o número de contornos
contornos, _= cv2.findContours(mask_hsv, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
# PARA DESENHAR O CONTORNO PRIMEIRO FAZ UMA CÓPIA DA IMAGEM


mask_rgb = cv2.cvtColor(mask_hsv,cv2.COLOR_GRAY2RGB)
contornos_img = mask_rgb.copy() #Copia da máscara para ser desenhada "por cima"

#-1 - referenciando um eixo que não existe para exibir todos os pontos encontrados
#[255,0,0] - rgb
#5 - espessura do traço
cv2.drawContours(contornos_img, contornos, -1, [255,0,0], 5);

plt.figure(figsize=(8,6))
plt.imshow(contornos_img);
```

```python
# notamos que a função retorna um dicionario

cnt= contornos[0]

M= cv2.moments(cnt)
print(M)
```

```python
#Calculo das coordenadas do centro da massa

cx = int(M['m10']/M["m00"])
cy = int(M['m01']/M["m00"])

print("centro de massa possição: ", cx, cy)
```

```python
## para desenhar a cruz vamos passar a cor e o tamanho em pixel
size = 20
color = (128,128,0)


cv2.line(contornos_img,(cx - size,cy),(cx + size,cy),color,5)
cv2.line(contornos_img,(cx,cy - size),(cx, cy + size),color,5)

# Para escrever vamos definir uma fonte

font = cv2.FONT_HERSHEY_SIMPLEX
text = cy , cx
origem = (0,50)

cv2.putText(contornos_img, str(text), origem, font,1,(200,50,0),2,cv2.LINE_AA)


plt.imshow(contornos_img);
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carregue a imagem
img = cv2.imread('bolinha.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Definição dos valores minimo e max da mascara
# o magenta tem h=300 mais ou menos ou 150 para a OpenCV
image_lower_hsv = np.array([140, 50, 100])
image_upper_hsv = np.array([170, 255, 255])
mask_hsv = cv2.inRange(img_hsv, image_lower_hsv, image_upper_hsv)


#contorno
contornos, _ = cv2.findContours(mask_hsv, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

mask_rgb = cv2.cvtColor(mask_hsv, cv2.COLOR_GRAY2RGB)
contornos_img = mask_rgb.copy() # Cópia da máscara para ser desenhada "por cima"

cv2.drawContours(contornos_img, contornos, -1, [255, 0, 0], 5);

plt.figure(figsize=(8,6))
plt.imshow(contornos_img)
```

```python

# Aplicar a máscara na imagem RGB
nova_img_rgb = cv2.bitwise_and(img_rgb, mask_rgb)

# Salvar a nova imagem modificada
cv2.imwrite('nova_imagem.png', nova_img_rgb)


plt.figure(figsize=(8,6))
plt.imshow(nova_img_rgb)
```
