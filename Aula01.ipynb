```python
#biblioteca OpenCV
#trabalhar imagens
import cv2

#import a biblioteca Numpy8 bits
#trabalhar matemática
import numpy as np

#linha magica para imprimir graficos no notebook
# Matplotlib é uma biblioteca da linguagem de programação Python,
#utilizada para visualização de dados e plotagem gráfica
%matplotlib inline
from matplotlib import pyplot as plt

#visualizar versão instalada
print ("OpenCV Versão : %s " % cv2.__version__)

```

```python
#exibindo imagem
#carregar img antes e confirmar o nome para exibir
image = cv2.imread("NATUREZA_1.jpg")
#plotagem da imagem
plt.imshow(image)
```

```python
#é necessário coverter de BGR para RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()

#opcionalmente platar a imagem sem o eixo
#para nao imprimir os eixos
image = cv2.imread("NATUREZA_1.jpg")
plt.axis('off')
plt.imshow(image)

#Mostrando a representação interna da imagem
print("Dimensões da imagem: ", image_rgb.shape)
print("Quantidade de linhas: ", image_rgb.shape[0])
print("Quantidade de colunas: ", image_rgb.shape[1])
print("Camadas de cores: ", image_rgb.shape[2])

# Mostrando a representação interna da imagem
print("Dimensões da imagem: \n", image_rgb)


#exibindo valores do código rgb da imagem especifico 50x50
(b, g, r) = image[50, 50]
print('O pixel (50, 50) tem as seguintes cores:')
print('Vermelho:', r, 'Verde:', g, 'Azul:', b)
```

```python
#Desafio
#plote suas componentes externas (shape) e internas (matriz)
image2 = cv2.imread("img3x3.png")
plt.imshow(image2)

#BGR -> RGB
image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
plt.show()

#Externas dimensões
print("Dimensões da imagem: ", image2_rgb.shape)
print("Quantidade de linhas: ", image2_rgb.shape[0])
print("Quantidade de colunas: ", image2_rgb.shape[1])
print("Camadas de cores: ", image2_rgb.shape[2])

#Internas matriz
print("Dimensões da imagem: \n", image2_rgb)

#plotando em cinza
# Carregando a imagem na versão tons de cinza (grayscale) de um arquivo
#imagem_cinza = cv2.imread("img3x3.png", cv2.IMREAD_GRAYSCALE)

# ou use o argumento 0, tem o mesmo efeito de importar na escala de cinza
#É necessário indicar que está em escala cinza
imagem_cinza = cv2.imread("img3x3.png", 0)
plt.imshow(imagem_cinza, cmap="gray")
plt.axis('off')
plt.show()
#exibindo matriz (claro =255 e escuro= 0)
imagem_cinza

# O redimensionamento da imagem pode ser feito na OpenCV
# através do comando cv2.resize(imagem, tamanho, interpolação)
# O tamanho é dado por uma tupla (W,H), onde W é a largura
# (número de colunas) e H é a altura (número de linhas)
# Carregando a imagem na versão colorida de um arquivo
#repetição de plotagem, conversão brg para rgb e print das dimesões
import cv2
import matplotlib.pyplot as plt
imagem = cv2.imread("NATUREZA_1.jpg")
image = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
print("Dimensões da imagem: ", image.shape)
#alterando tamanho
imagem2 = cv2.resize(image, (600,400), cv2.INTER_LINEAR)
print("Novas dimensões da imagem: ", imagem2.shape)
plt.imshow(imagem2)
plt.show()
```

```python
#Desafio 03
#implementar com a escala magenta
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagem3 = cv2.imread("img3x3.png")
imagemNova = cv2.cvtColor(imagem3, cv2.COLOR_BGR2RGB)

plt.imshow(imagemNova)
plt.show()

# Alterar o pixel
imagemNova[0, 0] = (255, 0, 255)
imagemNova[0, 1] = (0, 0, 0)
imagemNova[0, 2] = (255, 0, 0)

# Exibir a imagem com o pixel alterado
plt.imshow(imagemNova)
plt.show()
```
