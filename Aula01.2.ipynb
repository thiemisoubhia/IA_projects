```python
import cv2
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("NATUREZA_1.jpg")
plt.axis("off")
plt.imshow(image)
plt.show()
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(image_rgb)
plt.show()
```

```python
#informacoes da imagem = Linhas, colunas e camadas
print("Imagem real", image_rgb.shape)
print("Linhas", image_rgb.shape[0])
print("Colunas", image_rgb.shape[1])
print("Camadas", image_rgb.shape[2])
```

```python
#exibir matrizes
print("Matriz \n", image_rgb)
```

```python
#plotar mais de uma imagem simultaneamente
plt.figure(figsize=(10,10))
#2,2  o tamanho
#1  a posio
plt.subplot(2,2,1), plt.imshow(image_rgb)
plt.subplot(2,2,2), plt.imshow(image_rgb)
plt.subplot(2,2,3), plt.imshow(image_rgb)
plt.subplot(2,2,4), plt.imshow(image_rgb)
```

```python
#alterando tamanho
imagem2 = cv2.resize(image, (600,400), cv2.INTER_LINEAR)
print("Novas dimenses da imagem: ", imagem2.shape)
plt.imshow(imagem2)
plt.show()
```

```python
image2 = cv2.imread("img3x3.png")
plt.imshow(image2)
```

```python
#bgr para rgb
image2_rgb2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
plt.imshow(image2_rgb2)
```

```python
print("Imagem real", image2_rgb2.shape)
print("Linhas", image2_rgb2.shape[0])
print("Colunas", image2_rgb2.shape[1])
print("Camadas", image2_rgb2.shape[2])
```

```python
print("Matriz \n", image2_rgb2)
```

```python
#escala em cinza
imagem_cinza = cv2.imread("img3x3.png", 0)
plt.imshow(imagem_cinza, cmap="gray")
plt.axis('off')
plt.show()
```

```python
imagemNova = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
plt.imshow(imagemNova)
```

```python
#alterando o pixel
imagemNova[0,0] = [255,0,255]
imagemNova[0,1] = [0,0,0]
imagemNova[0,2] = [25,205,125]

plt.imshow(imagemNova)
```

# Array de ZERO

```python
#criar array com numpy
zeros_array = np.zeros((5,5))
zeros_array[0,0] = 1
plt.imshow(zeros_array, cmap="gray")
print(zeros_array)
```

```python
img = np.zeros((3,3))

#alterando pixel
img[0,0] = 80

#definindo valor minimo e maximo para alterar as cores na escala cinza
plt.imshow(img, cmap="gray", vmin=0, vmax=255)
plt.colorbar()
```
