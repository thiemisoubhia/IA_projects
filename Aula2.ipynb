```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import math
```

```python
img_BGR = cv2.imread("Folha.jpg")
img_RGB = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB)
plt.imshow(img_RGB)
```

```python
R, G, B = cv2.split(img_RGB)
plt.imshow(R, cmap="gray")
plt.show()

plt.imshow(G, cmap="gray")
plt.show()

plt.imshow(B, cmap="gray")
plt.show()
```

```python
num_lin = img_RGB.shape[0]
num_col = img_RGB.shape[1]

#Inteiro em 8 bits para a escala cinza
img_gray = np.zeros((num_lin, num_col))

#fazendo a varredura de uma imagem
for l in range (num_lin):
  for c in range (num_col):
    (r, g, b) = img_RGB[l,c]
    img_gray[l,c] = (int(r)+ int(g)+ int(b))/3
plt.imshow(img_gray, cmap= "gray")
```

```python
img_gray = cv2.cvtColor(img_RGB, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray, cmap="gray")
```

```python
img_cor = cv2.imread("Flor.jpg")
B, G, R= cv2.split(img_cor)
img_cinza = cv2.cvtColor(img_cor, cv2.COLOR_RGB2GRAY)
plt.imshow(img_cinza, cmap="gray")
#limiar= identificação de um tom da imagem, definindo um objeto
thresh = 50
[thresh, img_bin] = cv2.threshold(R, thresh, 255, cv2.THRESH_OTSU)
plt.imshow(img_cor)
plt.show()

plt.imshow(img_bin, cmap="gray")
```

```python
#inverter as cores, escala negativa
img = cv2.imread("Flor.jpg")
e= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
s = e.max() - e
plt.imshow(e, cmap="gray")
plt.show()
plt.imshow(s, cmap="gray")
```

```python
img_cogumelo = cv2.imread("cogumelo.png")
image = cv2.cvtColor(img_cogumelo, cv2.COLOR_BGR2GRAY)

for y in range(0, image.shape[0]):
  for x in range(0, image.shape[1]):
     if image[y, x] == 255:
      image[y, x] = 0
     else:
      image[y, x] = 255

plt.imshow(image, cmap= "gray")
```
