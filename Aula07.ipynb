```python
import cv2
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
import math
```

```python
img= cv2.imread("goku.jpg")
img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#cria uma matriz com o mesmo dimensional da imagem original, mas com valores 100
matriz = np.ones(img.shape, dtype="uint8")*150

img2 = cv2.add(img, matriz)

plt.imshow(img)
plt.show()
plt.imshow(img2)
```

```python
src1 = cv2.imread("lena.jpg")
scr1_rgb = cv2.cvtColor(src1, cv2.COLOR_BGR2RGB)

src2 = cv2.imread("rocket.jpg")
scr2_rgb = cv2.cvtColor(src2, cv2.COLOR_BGR2RGB)


# 50% transparecia para cada image
dst = cv2.addWeighted(scr1_rgb, 0.5, scr2_rgb, 0.5, 0)

plt.imshow(dst)
plt.show()
```

```python
img = cv2.imread("lena.jpg", 0)

plt.imshow(img, cmap="gray")
plt.show()

#Aplicando NOT
m_not = cv2.bitwise_not(img)

plt.imshow(m_not, cmap="gray")
plt.show()
```

```python
#Criando uma imagem inicial preta
mask= np.zeros((img.shape[0], img.shape[1]), dtype="uint8")

#coordenada central
center = (int(img.shape[1]/2),int(img.shape[0]/2))

#tamanho do raio
radius = int((img.shape[0]/2)*0.8)

#desenhando o circulo branco
cv2.circle(mask, center, radius, 255, -1)

plt.imshow(mask, cmap="gray")
plt.show()
```

```python
#aplicando AND
m_and = cv2.bitwise_and(img, mask)
plt.imshow(m_and, cmap="gray")
plt.show()
```

```python
#aplicando OR
m_or = cv2.bitwise_or(img, mask)
plt.imshow(m_or, cmap="gray")
plt.show()
```

```python
#aplicando XOR
m_xor = cv2.bitwise_xor(img, mask)
plt.imshow(m_xor, cmap="gray")
plt.show()
```
